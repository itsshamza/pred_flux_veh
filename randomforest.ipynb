```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import joblib
import matplotlib.pyplot as plt

data = pd.read_csv('data_entrante.csv')

data['Datetime'] = pd.to_datetime(data['Datetime'])

# Extraction du jour du mois depuis la colonne 'Datetime' et ajout de la colonne 'DayOfMonth'
data['DayOfMonth'] = data['Datetime'].dt.day


features = ['DayOfWeek', 'Hour','DayOfMonth','temperature(degC)', 'pression(hPa)', 'humidite(%)', 
            'point_de_rosee(degC)', 'vent_moyen(km/h)', 'vent_direction(deg)', 
            'pluie_3h(mm)', 'pluie_1h(mm)']

# features = ['Hour','DayOfWeek', 'DayOfMonth','Month', 'temperature(degC)', 'point_de_rosee(degC)', 
#                'humidite(%)', 'precipitations(mm)', 'neige(mm)', 'vent_direction(deg)', 
#                'vent_moyen(km/h)', 'rafale_vent_max(km/h)', 'pression(hPa)']


target_columns = [col for col in data.columns if col.startswith(('VL', 'PL', '2R'))]

# Paramètres à tester dans GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],          # Nombre d'arbres dans la forêt
    'max_depth': [10, 20, 30],               # Profondeur maximale de l'arbre
    'min_samples_split': [2, 5, 10],         # Minimum d'échantillons nécessaires pour diviser un nœud
    'min_samples_leaf': [1, 2, 4],           # Minimum d'échantillons nécessaires pour un nœud feuille
    'bootstrap': [True, False]               # Méthode d'échantillonnage
}

# Création des dossiers pour sauvegarder les modèles et les scalers
models_folder = "./randomforest_dataentrante_50"
os.makedirs(models_folder, exist_ok=True)

# Variables pour accumuler les résultats globaux
total_accuracy = 0
total_models = len(target_columns)
feature_importance_accumulator = np.zeros(len(features))
X = data[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Sauvegarde du scaler normalisé
scaler_filename = 'scaler_features_random_data_entrante.pkl'
joblib.dump(scaler, scaler_filename)
print(f"Scaler saved as {scaler_filename}")

# Boucle à travers chaque colonne cible
for vehicle_column in target_columns:
    print(f"Training model for target column: {vehicle_column}")
    
    # Créer la colonne de classe en divisant les valeurs des flux par 50 (pour la classification)
    data['Classe'] = (data[vehicle_column] // 50).astype('Int64')

    # Séparation des données en target (la variable y)
    y = data['Classe']

    # Diviser les données normalisées en ensemble d'entraînement et de test
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    # Initialisation du modèle RandomForestClassifier
    rf_model = RandomForestClassifier(random_state=42)

    # Recherche des meilleurs hyperparamètres avec GridSearchCV
    grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                               cv=3, n_jobs=-1, verbose=2, scoring='accuracy')

    grid_search.fit(X_train, y_train)


    best_rf = grid_search.best_estimator_

    model_filename = os.path.join(models_folder, f'random_forest_{vehicle_column}.joblib')
    joblib.dump(best_rf, model_filename)
    print(f"Model saved for {vehicle_column} as {model_filename}")

    y_pred = best_rf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    total_accuracy += accuracy
    feature_importance_accumulator += best_rf.feature_importances_

    print(f"\nResults for {vehicle_column}:")
    print(f"Best Parameters: {grid_search.best_params_}")
    print(f"Accuracy: {accuracy}")
    print(f"Classification Report for {vehicle_column}:")
    print(classification_report(y_test, y_pred))

# Calcul de l'importance moyenne des features sur tous les modèles
average_feature_importances = feature_importance_accumulator / total_models

# Afficher l'importance moyenne des features
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': average_feature_importances
}).sort_values(by='Importance', ascending=False)

print("\nAverage Feature Importance Across All Models:")
print(importance_df)

plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='green')
plt.title('Average Feature Importance Across All Models')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.gca().invert_yaxis()
plt.show()

# Calculer l'accuracy générale sur tous les modèles
average_accuracy = total_accuracy / total_models
print(f"\nAverage Accuracy across all models: {average_accuracy}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates
import os
import joblib
from sklearn.preprocessing import StandardScaler

days_of_week = {0: 'Lundi', 1: 'Mardi', 2: 'Mercredi', 3: 'Jeudi', 4: 'Vendredi', 5: 'Samedi', 6: 'Dimanche'}
colors = {0: 'blue', 1: 'green', 2: 'orange', 3: 'red', 4: 'purple', 5: 'brown', 6: 'pink'}

def predict_with_random_forest(models_folder, data, scaler_path, required_columns):
    """
    Fait des prédictions avec les modèles Random Forest pour chaque colonne cible.
    """
    # Charger le scaler utilisé pour les features
    scaler = joblib.load(scaler_path)
    
    # Normaliser les données d'entrée
    data_scaled = pd.DataFrame(scaler.transform(data[required_columns]), columns=required_columns)
    data_scaled['Datetime'] = data['Datetime']

    # Initialiser un DataFrame pour stocker les prédictions
    predictions_df = pd.DataFrame(index=data.index)
    predictions_df['Datetime'] = data['Datetime']

    for model_filename in os.listdir(models_folder):
        if model_filename.endswith('.joblib'):
            column_name = model_filename.replace('random_forest_', '').replace('.joblib', '')
            
            model = joblib.load(os.path.join(models_folder, model_filename))
            predictions = model.predict(data_scaled[required_columns])

            predictions_df[f'random_forest_{column_name}'] = predictions

    return predictions_df

def plot_regression_predictions_vs_history(predictions, previous_data=None, pattern=None, days=4):
    if pattern is None:
        print("Veuillez spécifier un motif pour filtrer les colonnes.")
        return
    
    columns_to_plot = [col for col in predictions.columns if pattern in col]
    if len(columns_to_plot) == 0:
        print(f"Aucune colonne trouvée avec le motif '{pattern}' dans les données de prédiction.")
        return

    predictions = predictions.loc[:, ~predictions.columns.duplicated()]
    if previous_data is not None:
        previous_data = previous_data.loc[:, ~previous_data.columns.duplicated()]

    try:
        predictions['Datetime'] = pd.to_datetime(predictions['Datetime'], errors='coerce')
        if previous_data is not None:
            previous_data['Datetime'] = pd.to_datetime(previous_data['Datetime'], errors='coerce')
    except Exception as e:
        print(f"Erreur lors de la conversion de 'Datetime': {e}")
        return

    last_days = predictions['Datetime'].max() - pd.Timedelta(days=days)
    predictions = predictions[predictions['Datetime'] >= last_days]
    if previous_data is not None:
        previous_data = previous_data[previous_data['Datetime'] >= last_days]

    predictions['DayOfWeek'] = predictions['Datetime'].dt.weekday.map(days_of_week)
    if previous_data is not None:
        previous_data['DayOfWeek'] = previous_data['Datetime'].dt.weekday.map(days_of_week)

    if predictions['Datetime'].isna().sum() > 0:
        print(f"Attention : {predictions['Datetime'].isna().sum()} valeurs 'Datetime' non valides détectées dans les prédictions.")
    if previous_data is not None and previous_data['Datetime'].isna().sum() > 0:
        print(f"Attention : {previous_data['Datetime'].isna().sum()} valeurs 'Datetime' non valides détectées dans les données historiques.")

    plt.figure(figsize=(14, 8))

    if previous_data is not None:
        for col in columns_to_plot:
            historical_col = col.replace("random_forest_", "")  # Ajustement pour les colonnes historiques
            if historical_col in previous_data.columns:
                # Diviser les valeurs historiques par 50 pour afficher les classes
                previous_data_class = (previous_data[historical_col] // 50).astype(int)
                plt.scatter(previous_data['Datetime'], previous_data_class, 
                            c=previous_data['Datetime'].dt.weekday.map(colors), 
                            label=f'{historical_col} (Historique - Classes)', alpha=0.6)

    for col in columns_to_plot:
        plt.scatter(predictions['Datetime'], predictions[col], 
                    c=predictions['Datetime'].dt.weekday.map(colors), 
                    label=f'{col} (Prédiction)', marker='x')

    ax = plt.gca()
    
    # Ajustement pour afficher les heures de façon plus espacée
    ax.xaxis.set_major_locator(mdates.HourLocator(interval=6))  # Afficher chaque 6 heures
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m %Hh'))  # Format jour et heure

    plt.xticks(rotation=45, fontsize=10)  # Ajuster la rotation et la taille des labels pour lisibilité
    plt.xlabel('Jour et Heure')
    plt.ylabel('Classe de Flux (par 50 véhicules)')
    plt.title(f"Flux des colonnes contenant '{pattern}' (Prédictions vs Historique sur {days} jours)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()  # Ajustement pour éviter le chevauchement des labels
    plt.show()

# Exemple d'utilisation
df = pd.read_csv('extended_meteo_data_en.csv')
required_columns = ['Hour','DayOfWeek', 'DayOfMonth','Month', 'temperature(degC)', 'point_de_rosee(degC)', 
               'humidite(%)', 'precipitations(mm)', 'neige(mm)', 'vent_direction(deg)', 
               'vent_moyen(km/h)', 'rafale_vent_max(km/h)', 'pression(hPa)']


# Charger et faire des prédictions avec les modèles Random Forest
predictions = predict_with_random_forest(models_folder='./randomforest_1.5month_50', data=df, scaler_path='./scaler_features_random_1.5month.pkl', required_columns=required_columns)

# Afficher les résultats
plot_regression_predictions_vs_history(predictions, previous_data=df, pattern='VL-P0', days=15)
```

```python
import os
import pandas as pd
import joblib
from meteostat import Point, Hourly
from sklearn.preprocessing import StandardScaler
import numpy as np


def predict_vehicle_flow(models_folder, location, prediction_dates):

    start = prediction_dates[0]
    end = prediction_dates[-1]
    meteo_data = Hourly(location, start, end).fetch()

    meteo_data['DayOfWeek'] = meteo_data.index.dayofweek
    meteo_data['Hour'] = meteo_data.index.hour
    meteo_data['DayOfMonth'] = meteo_data.index.day
    meteo_data['Month'] = meteo_data.index.month
    
    meteo_data.rename(columns={
        'temp': 'temperature(degC)',
        'dwpt': 'point_de_rosee(degC)',
        'rhum': 'humidite(%)',
        'prcp': 'precipitations(mm)',
        'snow': 'neige(mm)',
        'wdir': 'vent_direction(deg)',
        'wspd': 'vent_moyen(km/h)',
        'wpgt': 'rafale_vent_max(km/h)',
        'pres': 'pression(hPa)',
        'tsun': 'ensoleillement(H)'
    }, inplace=True)

    meteo_data['Datetime'] = meteo_data.index

    features = ['DayOfWeek', 'Hour', 'DayOfMonth', 'temperature(degC)', 
                'point_de_rosee(degC)', 'humidite(%)', 'precipitations(mm)', 
                'neige(mm)', 'vent_direction(deg)', 'vent_moyen(km/h)', 
                'rafale_vent_max(km/h)', 'pression(hPa)']

    meteo_data.fillna(0, inplace=True)  # Remplacer les valeurs manquantes par 0


    predictions = {}
    scaler = joblib.load('scaler_randomforest.pkl')
    scaled_features = scaler.transform(meteo_data[features])

    target_columns = [col.replace(".joblib", "") for col in os.listdir(models_folder) if col.endswith(".joblib")]

    for vehicle_column in target_columns:
        model_path = os.path.join(models_folder, f"{vehicle_column}.joblib")
        
        if os.path.exists(model_path):
            model = joblib.load(model_path)
            

            y_pred = model.predict(scaled_features)
            
            predictions[vehicle_column] = y_pred
            print(f"Predictions for {vehicle_column}: {y_pred[:10]}")  # Afficher les 10 premières prédictions
        else:
            print(f"Model for {vehicle_column} not found in {models_folder}")

    prediction_df = pd.DataFrame(predictions)
    prediction_df['Datetime'] = meteo_data['Datetime']
    
    return prediction_df


models_folder = '3month_random_class50'  
location = Point(44.8378, -0.5792)  # Coordonnées du point météo
prediction_dates = pd.date_range('2022-10-04', '2023-01-17', freq='H')  # Période de prédiction


prediction_df = predict_vehicle_flow(models_folder, location, prediction_dates)

prediction_df['Datetime'] = prediction_dates[:len(prediction_df)]


print(prediction_df)
prediction_df.to_csv('predictions.csv')
```

```python
import os
import joblib
import pandas as pd
from sklearn.metrics import accuracy_score

def calculate_general_accuracy(models_folder, X_test, y_test, target_columns):
    """
    Cette fonction charge les modèles depuis un dossier, prédit les valeurs pour chaque colonne cible 
    et renvoie l'accuracy générale.

    :param models_folder: Dossier contenant les modèles enregistrés
    :param X_test: Features d'entrée de test
    :param y_test: Cible réelle de test (DataFrame avec chaque colonne cible comme clé et valeurs correspondantes)
    :param target_columns: Liste des colonnes cibles pour lesquelles les modèles ont été entraînés
    :return: Accuracy générale
    """
    total_accuracy = 0
    total_models = len(target_columns)
    
    for column in target_columns:
        model_path = os.path.join(models_folder, f'random_forest_{column}.joblib')
        
        if os.path.exists(model_path):
            model = joblib.load(model_path)
            
            y_pred = model.predict(X_test)
            y_test[column]= y_test[column]//50

            accuracy = accuracy_score(y_test[column], y_pred)
            total_accuracy += accuracy
            print(f"Accuracy for {column}: {accuracy}")
        else:
            print(f"Model for {column} not found in {models_folder}")
    

    general_accuracy = total_accuracy / total_models if total_models > 0 else 0
    print(f"General Accuracy: {general_accuracy}")
    
    return general_accuracy


data = pd.read_csv('data_entrante.csv')

features = ['DayOfWeek', 'Hour', 'temperature(degC)', 'pression(hPa)', 'humidite(%)', 
            'point_de_rosee(degC)', 'vent_moyen(km/h)', 'vent_direction(deg)', 
            'pluie_3h(mm)', 'pluie_1h(mm)']


target_columns = [col for col in data.columns if col.startswith(('VL', 'PL', '2R'))]

last_week = data[:68]  

X_test_last_week = last_week[features]  # Sélectionner uniquement les colonnes de caractéristiques
y_test_last_week = last_week[target_columns]  # Sélectionner les colonnes cibles

# Dossier des modèles
models_folder = r"C:\Users\Admin\Desktop\données_V1\random_class50"

# Appel de la fonction pour calculer l'accuracy générale
general_accuracy = calculate_general_accuracy(models_folder, X_test_last_week, y_test_last_week, target_columns)

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import xgboost as xgb
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Étape 1: Charger les données depuis le fichier CSV
data = pd.read_csv('data_entrante.csv')

features = ['DayOfWeek', 'Hour', 'temperature(degC)', 'pression(hPa)', 'humidite(%)', 
            'point_de_rosee(degC)', 'vent_moyen(km/h)', 'vent_direction(deg)', 
            'pluie_3h(mm)', 'pluie_1h(mm)']

vehicle_column = 'VL-P25'


max_value = data[vehicle_column].max()

# Calculer la taille de chaque classe
class_size = max_value / 5

# Diviser les valeurs par la taille de la classe pour obtenir les classes, et arrondir
data['Classe'] = (data[vehicle_column] // class_size).astype('Int64')

# Remapper les classes pour qu'elles soient continues
label_encoder = LabelEncoder()
data['Classe'] = label_encoder.fit_transform(data['Classe'])

# Étape 3: Séparer les données en features et cible
X = data[features]
y = data['Classe']

# Appliquer un scaler sur les données d'entraînement et de test
scaler = StandardScaler()

# Diviser les données en jeu d'entraînement et jeu de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Ajuster le scaler sur les données d'entraînement et transformer les données
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialiser et entraîner le modèle XGBoost Classifier
model = xgb.XGBClassifier(n_estimators=1000, learning_rate=0.1, objective='multi:softmax')
model.fit(X_train, y_train)

# Prédire sur l'ensemble de test
y_pred = model.predict(X_test)

# Remapper les prédictions et les vraies valeurs à leurs classes originales
y_test = label_encoder.inverse_transform(y_test)
y_pred = label_encoder.inverse_transform(y_pred)

# Évaluer le modèle
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Calculer les métriques de classification
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
from sklearn.model_selection import train_test_split

# Lire les données
data = pd.read_csv('data_entrante.csv', parse_dates=['Datetime'])

# List of posts to predict
posts = ['2R-P01', '2R-P02', '2R-P03', '2R-P04', '2R-P05', '2R-P06', '2R-P08', '2R-P09', '2R-P10', '2R-P12', '2R-P13', '2R-P14', 
         '2R-P17', '2R-P18', '2R-P19', '2R-P20', '2R-P21', '2R-P23', '2R-P24', '2R-P25', '2R-P26', 'PL-P01', 'PL-P02', 
         'PL-P03', 'PL-P04', 'PL-P05', 'PL-P06', 'PL-P08', 'PL-P09', 'PL-P10', 'PL-P12', 'PL-P13', 'PL-P14', 'PL-P17', 
         'PL-P18', 'PL-P19', 'PL-P20', 'PL-P21', 'PL-P23', 'PL-P24', 'PL-P25', 'PL-P26', 'VL-P01', 'VL-P02', 'VL-P03', 
         'VL-P04', 'VL-P05', 'VL-P06', 'VL-P08', 'VL-P09', 'VL-P10', 'VL-P12', 'VL-P13', 'VL-P14', 'VL-P17', 'VL-P18', 
         'VL-P19', 'VL-P20', 'VL-P21', 'VL-P23', 'VL-P24', 'VL-P25', 'VL-P26']

# Normaliser les données d'entrée
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data.drop(['Datetime'], axis=1))

# Fonction pour créer des séquences
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data)):
        end_ix = i + n_steps
        if end_ix > len(data)-1:
            break
        seq_x, seq_y = data[i:end_ix, :-len(posts)], data[end_ix, -len(posts):]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

# Définir le nombre de pas de temps
n_steps = 10

# Créer les séquences
X, y = create_sequences(data_scaled, n_steps)

# Diviser les données en ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Construire le modèle RNN
def build_rnn(input_shape):
    model = Sequential()
    model.add(SimpleRNN(50, activation='relu', input_shape=input_shape))
    model.add(Dense(len(posts)))
    model.compile(optimizer='adam', loss='mse')
    return model

# Initialiser et entraîner le modèle
rnn_model = build_rnn((X_train.shape[1], X_train.shape[2]))
rnn_model.fit(X_train, y_train, epochs=20, verbose=1, validation_data=(X_test, y_test))

# Évaluer le modèle
loss = rnn_model.evaluate(X_test, y_test)
print(f"Loss: {loss}")

# Prédiction
y_pred = rnn_model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error

# Prédiction
y_pred = rnn_model.predict(X_test)

# Reshape des données pour qu'elles aient la même forme
y_test_reshaped = y_test.reshape(-1, y_test.shape[-1])
y_pred_reshaped = y_pred.reshape(-1, y_pred.shape[-1])

# Calcul du MSE
mse = mean_squared_error(y_test_reshaped, y_pred_reshaped)
print(f"Mean Squared Error (MSE): {mse}")

# Calcul du R²
r2 = r2_score(y_test_reshaped, y_pred_reshaped)
print(f"Coefficient of Determination (R²): {r2}")

# Calcul du MAPE
mape = mean_absolute_percentage_error(y_test_reshaped, y_pred_reshaped)
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")

```
